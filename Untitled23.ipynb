```python
#Question 2: Check for missing values in all columns and describe your approach for handling them.

import pandas as pd

# Load dataset from GitHub raw link (update with your correct link)
url = "https://raw.githubusercontent.com/sonia652/bike.detail/refs/heads/main/BIKE%20DETAILS.csv"
df = pd.read_csv(url)

# Check missing values
print("Missing values in each column:\n")
print(df.isnull().sum())

# Handling approach:
# - For numerical columns: fill missing values with median
# - For categorical columns: fill missing values with mode
# - If a column has too many missing values, drop it

# Example implementation
for col in df.columns:
    if df[col].dtype == 'object':  # categorical
        df[col] = df[col].fillna(df[col].mode()[0])
    else:  # numerical
        df[col] = df[col].fillna(df[col].median())

# Verify no missing values remain
print("\nMissing values after handling:\n")
print(df.isnull().sum())
```

```python
#Question 3: Plot the distribution of selling prices using a histogram and describe the overall trend.


import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/sonia652/bike.detail/refs/heads/main/BIKE%20DETAILS.csv"
df = pd.read_csv(url)

# Plot histogram of selling prices
plt.figure(figsize=(8,5))
plt.hist(df["selling_price"], bins=20, edgecolor="black")
plt.xlabel("Selling Price")
plt.ylabel("Frequency")
plt.title("Distribution of Selling Prices")
plt.show()
```

```python
#Question 4: Create a bar plot to visualize the average selling price for each seller_type and write one observation.

import seaborn as sns

# Calculate average selling price per seller type
avg_price = df.groupby("seller_type")["selling_price"].mean().reset_index()

# Plot bar chart
plt.figure(figsize=(7,5))
sns.barplot(x="seller_type", y="selling_price", data=avg_price)
plt.xlabel("Seller Type")
plt.ylabel("Average Selling Price")
plt.title("Average Selling Price by Seller Type")
plt.show()
```

```python
#Question 5: Compute the average km_driven for each ownership type (1st owner,2nd owner, etc.), and present the result as a bar plot.


import seaborn as sns
import matplotlib.pyplot as plt

# Group by ownership type and calculate average km_driven
avg_km = df.groupby("owner")["km_driven"].mean().reset_index()

# Plot bar chart
plt.figure(figsize=(8,5))
sns.barplot(x="owner", y="km_driven", data=avg_km)
plt.xlabel("Ownership Type")
plt.ylabel("Average km Driven")
plt.title("Average km Driven by Ownership Type")
plt.xticks(rotation=45)
plt.show()
```

```python
#Question 6: Use the IQR method to detect and remove outliers from the km_driven column. Show before-and-after summary statistics.


# Before removing outliers
print("Before removing outliers:")
print(df["km_driven"].describe())

# IQR method
Q1 = df["km_driven"].quantile(0.25)
Q3 = df["km_driven"].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_no_outliers = df[(df["km_driven"] >= lower_bound) & (df["km_driven"] <= upper_bound)]

# After removing outliers
print("\nAfter removing outliers:")
print(df_no_outliers["km_driven"].describe())
```

```python
#Question 7: Create a scatter plot of year vs. selling_price to explore the relationship between a bike's age and its price.

plt.figure(figsize=(8,6))
sns.scatterplot(x="year", y="selling_price", data=df, alpha=0.6)
plt.xlabel("Year of Manufacture")
plt.ylabel("Selling Price")
plt.title("Scatter Plot: Year vs Selling Price")
plt.show()

```
